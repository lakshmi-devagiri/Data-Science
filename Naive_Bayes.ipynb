Here, I can see some missing values in the dataset . I replace them with mean value

```python
input.Age=input.Age.fillna(input.Age.mean())
```

```python
##Split the dataset into train and test
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(input,target,test_size=0.2)
```

Here, I have imported the Train_test_split Library and assign the variables X_train, X_test,y_train, Y-test assign this variable train_test_split Function with parameters (input, output, test-Size)

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

Gaussian NB is used when data distribution is normal .It is also called Bell-curve.

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

After training the model then will check the accuracy
